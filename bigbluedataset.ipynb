```python
import pandas as pd
```

```python
def import_data():

    df = pd.read_csv('employee_salary_dataset.csv')

    return df
```

```python
def preprocessing(df):
    
    #σβηνω τις null τιμες 
    df.dropna(inplace = True)

    #(ελεγχος για διπα)
    df.duplicated(subset=['EmployeeID']).sum()
    #κανω ενα drop το column name  γτ δεν προσφεερει κατι στην ανάλυση
    df.drop(columns='Name', inplace = True)


    #προσθετω κλιμακα στο salary
    categories = []
    for x in df['Monthly_Salary']:
        if x < 70000 :
            categories.append('LOW')
        elif x > 100000 :
            categories.append('HIGH')
        else: 
            categories.append('MEDIUM')
    
    df['Salary_Category'] = categories



    #δημιουργια seniority  αναλογα με τα χρονια 
    category_seniority= []
    for x in df['Experience_Years']:
        if x <=  2:
            category_seniority.append('ENTRY_LEVEL')
        elif x <= 5 :
            category_seniority.append('MID_LEVEL')
        else: 
            category_seniority.append('SENIOR_LEVEL')
    
    df['Seniority'] = category_seniority



    #μεσος ορος ανα department
    bydep = df.groupby('Department')['Monthly_Salary'].mean().round(2)
    
    df['Department_Average'] = df['Department'].map(bydep)

    return df
```
